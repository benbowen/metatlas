```python
myExperimentID = '542f304ff207836cd3c82bc1'
myArray = 'lcms_test_1'
# define lists of the files that you want to study as groups
blankFiles_pos = [10668, 10669, 10670] #extraction blank files positive mode
mediaFiles_pos = [10673,10671,10675] #media controls in positive mode
spentFiles_pos = [10672,10674,10676] #spent media files in positive mode
qc_CAS16_Files_pos = [10684] #casette_16 of standards run as QCs
qc_CAS17_Files_pos = [10686] #casette_17 of standards run as QCs
dictId = '5433367e762640542f846835'
url = 'https://metatlas.nersc.gov/api/dict/%s/' % dictId
r = client.get(url)
dictData = json.loads(r.text)
```

```python
metatlas = reload(metatlas)

# compound = dictData[u'compounds'][5] #leucine
compound = dictData[u'compounds'][9] #proline
# print compound
# data = metatlas.getEICForCompound(compound,myArray,qc_CAS17_Files_pos[0],0.5,client)
data = metatlas.getEICForCompound(compound,myArray,mediaFiles_pos[1],0.5,client)

fitResult = metatlas.fitACompound(compound,data)

fig, ax = plt.subplots(1, 1,figsize=(10,5))
metatlas.createChromatogramPlots(data,compound,fitResult,ax)

rtMax = fitResult[1]+2*fitResult[2]
rtMin = fitResult[1]-2*fitResult[3]
rtPeak = fitResult[1]
```

```python
payload = {'L':1,'P':1,'arrayname':myArray,'fileid':mediaFiles_pos[1],
           'max_rt': rtPeak+0.01,'min_rt':rtPeak-0.01,
           'queryType':'IvsMZinRTRange'}
url = 'https://metatlas.nersc.gov/api/run'
r = client.get(url,params=payload)
data = np.asarray(json.loads(r.content))
```

```python
ppm=20
mzMin = min(data[:,0])
mzMax = max(data[:,0])
f = np.ceil(1e6*np.log(mzMax/mzMin)/ppm)
mzEdges = np.logspace(np.log10(mzMin),np.log10(mzMax),f)
intensity =  np.interp(mzEdges,data[:,0],data[:,1],0,0)
plt.plot(mzEdges,intensity)
# plt.xlim(115,118)
```

```python
[maxPks,minPks]=metatlas.peakdet(intensity, 1e3, x = None)
```

```python
idx = np.argsort(-1*maxPks[:,1])
# print max(maxPks[ind,0])
idx = maxPks[idx,0]

mzPks = mzEdges[idx.astype(int)]
# print mzEdges[maxPks[idx[:9],0]]
print mzPks[:200]
```

```python
M = np.zeros((200,len(rtEdges)))
mzLabel = []             
for i in range(0,200):
    files_I_want = mediaFiles_pos
    target = mzPks[i]
    
    mzMax = target + target * 50. / 1e6
    mzMin = target - target * 50. / 1e6

    myList = ','.join(map(str, files_I_want))

    payload = {'L':1,'P':1,'arrayname':myArray,'fileidlist':myList,
              'max_mz':mzMax,'min_mz':mzMin,
              'min_rt':rtMin,'max_rt':rtMax,
              'nsteps':20000,'queryType':'XICofFile_mf'}
    url = 'https://metatlas.nersc.gov/api/run'
    r = client.get(url,params=payload)
    data = np.asarray(json.loads(r.content))

    # for myFile in files_I_want:
    myFile = mediaFiles_pos[1]
    x1 = data[:,0][(data[:,2]==myFile)]
    y1 = data[:,1][(data[:,2]==myFile)]
    idx = np.argsort(x1)
    M[i,:] = np.interp(rtEdges,x1[idx],y1[idx],0,0)
    tempStr = '%.4f & %.2f' % (target,sum(y1))
    mzLabel.append(tempStr)
```

```python
nM = M
for i in range(0,200):

    idx = np.argwhere(M[i,:] > 0)
    temp = M[i,idx]
    temp = temp[:,0]
    temp = temp - min(temp)
    temp = temp / max(temp)
    x = rtEdges[idx]
    x = x[:,0]
    temp=np.interp(rtEdges,x,temp,0,0)
    nM[i,:] = temp
    plt.plot(rtEdges,temp)
```

```python
from scipy.spatial.distance import pdist, squareform
from scipy.cluster.hierarchy import linkage, dendrogram

distanceMatrix = pdist(nM, metric = 'euclidean')
fig, ax = plt.subplots(1, 1,figsize=(5,30))

R = dendrogram( linkage(distanceMatrix, method='complete'), 
           color_threshold=0.1,
           labels = mzLabel,
           orientation = 'left',
           ax = ax)
```

```python
from scipy.spatial.distance import cdist
idx = np.argmin(abs(mzPks[:50]-116.07))
print idx
# 
distanceMatrix = cdist([nM[idx,:]],nM, 'correlation')
print distanceMatrix.shape
sx = np.argsort(distanceMatrix)[0]
fig, ax = plt.subplots(1, 1,figsize=(12,9))
temp = nM[sx[0],:]

f = metatlas.fitfunc(fitResult, rtEdges)
f = f / sum(f)
temp = np.convolve(temp,f,'same')
ax.plot(rtEdges,temp, linewidth=4, label='%5.3f %s' % (distanceMatrix[0,sx[0]],mzLabel[sx[0]]))

for i in range(1,10):
    temp = nM[sx[i],:]
    if i<3:
        lw = 4
    else:
        lw=1
    temp = np.convolve(temp,f,'same')
    ax.plot(rtEdges,temp, linewidth=lw,label='%5.3f %s' % (distanceMatrix[0,sx[i]],mzLabel[sx[i]]))
legend = ax.legend(loc='upper right', shadow=True)
```
